```python
# Creacion de tabla Fact_Coins_and_Categories en 5 pasos:

# 1 Creo Lista con todas las Coins
lista_coin_ids = mkt_Cap_Coins["id"].iloc[:50].tolist()
lista_coin_ids
```

```python
# 2 creo funcion para obtener todas las categorias e ids en una lista
def ids_categorias(lista_ids):
    dict = []
    for n in lista_ids:
        cat = cg.get_coin_by_id(n)["categories"]
        dict.append({n:cat})
    return dict
```

```python
# 3 paso todas las categorias e ids a una lista
df = ids_categorias(lista_coin_ids)
df
```

```python
# 4 creacion de dataframe
ids = []
categories = []

# Unpack the dictionaries and store data in the lists
for item in df:
    for key, value in item.items():
        ids.extend([key] * len(value))
        categories.extend(value)

# Create a DataFrame
df_coins_cat = pd.DataFrame({
    'id': ids,
    'categories': categories
})

print(df_coins_cat)
```

```python
# 5 To csv
df_coins_cat.to_csv("Dim_Coins_and_Categories.csv",index=False)
```

```python
from datetime import date
def days_dif(fecha):
    timestamp = date(1970,1,1)
    fecha = date(fecha)
    diferencia = fecha-timestamp
    return diferencia
```

```python
def days_difference_from_timestamp(input_date_str):
# input_date_str = "2020-12-31"  # Format: YYYY-MM-DD
    input_date = date.fromisoformat(input_date_str)
    timestamp = date(1970, 1, 1)
    difference = input_date - timestamp
    return difference.days
```

```python
# Tomo datos desde el 2021
days_difference_from_timestamp("2021-01-01")
```

```python
# Tomo datos de hoy 2023-8-19
days_difference_from_timestamp("2023-08-19")
```

```python
# Convert a date to a Unix timestamp
from datetime import datetime

date_str = "2020-01-01"
def convert_to_timestamp(date_str :str):
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    timestamp = int(date_obj.timestamp())
    print(timestamp)
```
